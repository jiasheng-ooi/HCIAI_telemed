```python
from google.colab import files
from google.colab import drive
import glob
import re
import pandas as pd

drive.mount('/content/drive')
%cd /content/drive/MyDrive/HCI AI Mini Project Data/Clean Transcripts

# Retrieve the file list
file_list = glob.glob("*.txt")

# Sort the file list alphabetically
file_list.sort()

# Create an ExcelWriter object
excel_writer = pd.ExcelWriter('/content/drive/MyDrive/HCI AI Mini Project Data/extracted_conversation_data.xlsx')

# Process each file
for file_path in file_list[:30]:  #You can choose to put all if you want, but I capped so it will not take too long
    with open(file_path, 'r') as file:
        text = file.read()
        # Process the text file as required
        # Perform sentiment analysis or any other NLP tasks
        # Remove lines starting with 'D:'
        text_without_d_lines = re.sub(r'^D[;:] .*?\n', '', text, flags=re.MULTILINE)

        # Extract lines starting with 'P:' (omitting the 'P:')
        conversation_lines = re.findall(r'P: (.*?)(?=P:|$)', text_without_d_lines, re.DOTALL)
        # Filter lines with less than three words
        conversation_list = [re.sub(r'^.*\n(P:.*?)$', r'\1', line.strip()) for line in conversation_lines if len(re.findall(r'\w+', line)) >= 3]

        # Create a DataFrame for the conversation data
        df = pd.DataFrame({'Conversation': conversation_list})

        # Save the DataFrame to a sheet in the ExcelWriter object
        sheet_name = file_path[:-4]  # Extract the file name without extension
        df.to_excel(excel_writer, sheet_name=sheet_name, index=False)

# Save the ExcelWriter object to a single Excel file
excel_writer.save()

```

```python
#Load Transformers Model
from transformers import RobertaTokenizerFast, TFRobertaForSequenceClassification, pipeline

tokenizer = RobertaTokenizerFast.from_pretrained("arpanghoshal/EmoRoBERTa")
model = TFRobertaForSequenceClassification.from_pretrained("arpanghoshal/EmoRoBERTa")

emotion = pipeline('sentiment-analysis',
                    model='arpanghoshal/EmoRoBERTa')
```

```python
#Segregating the possible emotions into these types, purely by personal perspective
emotion_list = ['admiration', 'amusement', 'anger', 'annoyance', 'approval', 'caring', 'confusion', 'curiosity', 'desire', 'disappointment', 'disapproval', 'disgust', 'embarrassment', 'excitement', 'fear', 'gratitude', 'grief', 'joy', 'love', 'nervousness', 'optimism', 'pride', 'realization', 'relief', 'remorse', 'sadness', 'surprise', 'neutral']
positive_emotions = ['admiration', 'amusement', 'caring', 'curiosity', 'excitement', 'gratitude', 'joy', 'love', 'optimism', 'realization', 'relief', 'surprise']
neutral_emotions = [ 'neutral','pride','approval']
negative_emotions = ['anger', 'annoyance', 'desire', 'disappointment', 'disapproval', 'disgust', 'embarrassment', 'fear', 'grief', 'nervousness', 'remorse', 'sadness','confustion']
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

# Specify the path to the Excel file
excel_file = '/content/drive/MyDrive/HCI AI Mini Project Data/extracted_conversation_data.xlsx'

# Load the Excel file
workbook = load_workbook(excel_file)

# Iterate through each sheet
for sheet_name in workbook.sheetnames:
    # Read the sheet into a DataFrame
    df = pd.read_excel(excel_file, sheet_name=sheet_name)

    # Evaluate each conversation and add the emotion column
    df['Emotion'] = df['Conversation'].apply(lambda text: emotion(text)[0]['label'])

    # Get the active sheet
    sheet = workbook[sheet_name]

    # Determine the column index to start writing the emotion and distress columns
    start_col = sheet.max_column + 1

    # Write 'Emotion' and 'Distress' in bold to the first row
    sheet.cell(row=1, column=start_col, value='Emotion').font = Font(bold=True)
    sheet.cell(row=1, column=start_col+1, value='Distress').font = Font(bold=True)

    # Write the emotion and distress columns to the sheet starting from index 2
    for i, emotion_label in enumerate(df['Emotion'], start=2):
        sheet.cell(row=i, column=start_col, value=emotion_label)
        #Assigning Value to each emotions
        if emotion_label in positive_emotions:
            distress_value = -1
        elif emotion_label in neutral_emotions:
            distress_value = 0
        elif emotion_label in negative_emotions:
            distress_value = 1
        else:
            distress_value = None

        sheet.cell(row=i, column=start_col+1, value=distress_value)

# Save the modified Excel file
workbook.save(excel_file)
```

## **Output**

```python
import pandas as pd
from tabulate import tabulate

# Specify the path to the Excel file
excel_file = '/content/drive/MyDrive/HCI AI Mini Project Data/extracted_conversation_data.xlsx'

# Load the Excel file
with pd.ExcelFile(excel_file) as xls:
    distress_sum_dict = {}  # Dictionary to store distress sum for each sheet

    for sheet_name in xls.sheet_names:
        # Read the sheet into a DataFrame
        df = pd.read_excel(excel_file, sheet_name=sheet_name)

        # Calculate the sum of 'Distress' column
        distress_sum = df['Distress'].sum()

        # Add the distress sum to the dictionary
        distress_sum_dict[sheet_name] = distress_sum

# Sort the distress sums in descending order
sorted_distress_sums = sorted(distress_sum_dict.items(), key=lambda x: x[1], reverse=True)

# Prepare the data for the table
table_data = []
for sheet_name, distress_sum in sorted_distress_sums:
#Thresholds for Distress
    if distress_sum > 15:
        distress_level = 'High'
    elif distress_sum < 5:
        distress_level = 'Low'
    else:
        distress_level = 'Normal'
    table_data.append([sheet_name, distress_level, distress_sum])

# Print the table with title
table_title = 'Distress Levels by Patient'
table_headers = ['Patient', 'Distress Level', 'Distress Sum']
table = tabulate(table_data, headers=table_headers, tablefmt='fancy_grid')
print(f"{table_title}\n{table}")
```
